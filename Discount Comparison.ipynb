```python
# create a list of all suppliers dataframes

supplier_list = [supplier_1, supplier_2, supplier_3, supplier_4, supplier_5,
                 supplier_6, supplier_7, supplier_8, supplier_9,
                 supplier_10, supplier_11, supplier_12]




# making a loop to merge the original dataframe with all suppliers dataframes and create a list of merged dataframes

merged_list=[]
for supplier in supplier_list:
    merged = original_df.merge(supplier, how='inner',on=['new Name','Price','conc'])
    merged_list.append(merged)




# Merging all dataframes in the merged_list into one final dataframe

final_df = merged_list[0]    
for df in merged_list[1:]:
    final_df = final_df.merge(df, how='outer')
    
    
    
    
# drop the 'new Name' and 'conc' columns as they are not needed in the final output

final_df.drop(['new Name','conc'], axis = 1, inplace = True)
```

```python
final_df.head(5)
```

```python
# history_df contains previous discount data

history_df = pd.read_excel(r'excel files/History data.xlsx')



# identify maximum discount for each unique product in the history data

history_max_discount = history_df.groupby('Product name')['Discount %'].max().reset_index()

history_max_discount.rename(columns={'Discount %': 'Max Discount %'}, inplace=True)



# unifying some columns name to be matched with final_df dataframe

history_max_discount.rename(columns={'Product name': 'Item'}, inplace=True)




# meging the history_max_discount dataframe with final_df to add the maximum discount percentage for each product

final_df = final_df.merge(history_max_discount, how='left',on='Item')




# highlight the maximum discount percentage across all suppliers for each unique product

df_final = final_df.style.highlight_max(subset =['supplier_1', 'supplier_2','supplier_3','supplier_4','supplier_5','supplier_6','supplier_7',
                                               'supplier_8','supplier_9','supplier_10','supplier_11','supplier_12','Max Discount %'],
                                      color = 'lightgreen', axis = 1)

df_final
```

```python
# ! pip install xlsxwriter

# Save the final DataFrame to an Excel file with formatting

with pd.ExcelWriter('Result/Best Discounts.xlsx', engine='xlsxwriter') as writer:
    df_final.to_excel(writer, sheet_name='Best Discounts', index=False)
```
